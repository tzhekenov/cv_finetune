```python
import os
import shutil

def categorize_function(filename):
    # This function determines the folder name based on the filename
    # Customize this part to suit your categorization needs
    # Example: Return everything before the first underscore
    return filename.split('(')[0]

def organize_images(source_dir):
    for filename in os.listdir(source_dir):
        if filename.endswith(".jpg"):  # Filter for JPEG images
            category = categorize_function(filename)
            target_dir = os.path.join("/workspaces/finetune/data_test_labels", category)
            if not os.path.exists(target_dir):
                os.makedirs(target_dir)
            shutil.move(os.path.join(target_dir, filename), os.path.join(target_dir, filename))
            print(f"Moved {filename} to {target_dir}/")

# Set the directory where your images are currently stored
source_directory = "/workspaces/finetune/data_test"
organize_images(source_directory)
```

```python
import os
import shutil

def move_folders_to_new_parent(source_dir, parent_folder_name):
    # Path to the new parent folder
    parent_folder_path = os.path.join(source_dir, parent_folder_name)

    # Create the parent folder if it does not exist
    if not os.path.exists(parent_folder_path):
        os.makedirs(parent_folder_path)
        print(f"Created new directory: {parent_folder_path}")

    # Iterate over each item in the source directory
    for item in os.listdir(source_dir):
        item_path = os.path.join(source_dir, item)
        
        # Check if the item is a directory and not the new parent folder
        if os.path.isdir(item_path) and item != parent_folder_name:
            # Move the directory to the new parent folder
            shutil.move(item_path, parent_folder_path)
            print(f"Moved {item_path} to {parent_folder_path}/")

# Set the directory where your images are currently stored
source_directory = "/workspaces/finetune/data_test"
new_parent_folder = "organized_categories"

move_folders_to_new_parent(source_directory, new_parent_folder)
```

```python
import zipfile
import os

# Define the path to the ZIP file and the extraction directory
zip_path = "/workspaces/finetune/data_558 2.zip"  # Replace with the actual path to your ZIP file
extraction_path = "/workspaces/finetune/test_data"  # Replace with your desired extraction path

# Create the extraction directory if it doesn't exist
os.makedirs(extraction_path, exist_ok=True)

# Unzip the file and rename all contents to lowercase
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

    # Iterate over all the files and folders in the extraction path and rename them
    for root, dirs, files in os.walk(extraction_path, topdown=False):
        # Rename files
        for name in files:
            original_path = os.path.join(root, name)
            new_path = os.path.join(root, name.lower())
            os.rename(original_path, new_path)

        # Rename directories
        for name in dirs:
            original_path = os.path.join(root, name)
            new_path = os.path.join(root, name.lower())
            os.rename(original_path, new_path)

print("Files have been extracted and all names have been converted to lowercase.")
```

```python
import os
import shutil

# Define the paths for the source directories and the destination directory
base_path = "/workspaces/finetune/test_data/data_558"
dissimilar_path = os.path.join(base_path, "dissimilar")
similar_path = os.path.join(base_path, "similar")
destination_path = "/workspaces/finetune/combined_classes"

# Create the destination directory if it doesn't exist
os.makedirs(destination_path, exist_ok=True)

# Function to copy class folders
def copy_classes(source_directory):
    for class_folder in os.listdir(source_directory):
        src_folder_path = os.path.join(source_directory, class_folder)
        dest_folder_path = os.path.join(destination_path, class_folder)

        # Check if the destination folder already exists; if not, copy the folder
        if not os.path.exists(dest_folder_path):
            shutil.copytree(src_folder_path, dest_folder_path)
        else:
            # If it exists, copy individual files to avoid overwriting the entire folder
            for file_name in os.listdir(src_folder_path):
                src_file_path = os.path.join(src_folder_path, file_name)
                dest_file_path = os.path.join(dest_folder_path, file_name)
                if not os.path.exists(dest_file_path):  # Avoid overwriting existing files
                    shutil.copy(src_file_path, dest_file_path)

# Copy class folders from both dissimilar and similar directories
copy_classes(dissimilar_path)
copy_classes(similar_path)

print("Class folders from 'dissimilar' and 'similar' have been combined in:", destination_path)
```

```python
import os
import json

# Define the paths to the data directory and the output JSON files
data_folder = '/workspaces/finetune/combined_classes copy'
query_json_path = '/workspaces/finetune/query_set.json'
gallery_json_path = '/workspaces/finetune/gallery_set.json'

# Initialize lists to hold query and gallery data
query_set = []
gallery_set = []

# Loop through each directory in the data folder
for class_folder in os.listdir(data_folder):
    class_path = os.path.join(data_folder, class_folder)
    
    if os.path.isdir(class_path):
        images = os.listdir(class_path)
        
        if len(images) == 1:
            # Only one image in this class, add it to the gallery
            gallery_set.append({
                "image_path": os.path.join(class_path, images[0]),
                "class": class_folder
            })
        elif len(images) > 1:
            # More than one image, add the first to the query set, rest to the gallery
            query_set.append({
                "image_path": os.path.join(class_path, images[0]),
                "class": class_folder
            })
            gallery_set.extend([{
                "image_path": os.path.join(class_path, img),
                "class": class_folder
            } for img in images[1:]])

# Save the query and gallery data to JSON files
def save_to_json(data, file_path):
    with open(file_path, 'w') as f:
        json.dump(data, f, indent=4)

save_to_json(query_set, query_json_path)
save_to_json(gallery_set, gallery_json_path)
```
