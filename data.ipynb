```python
import os
import shutil

def categorize_function(filename):
    # This function determines the folder name based on the filename
    # Customize this part to suit your categorization needs
    # Example: Return everything before the first underscore
    return filename.split('(')[0]

def organize_images(source_dir):
    for filename in os.listdir(source_dir):
        if filename.endswith(".jpg"):  # Filter for JPEG images
            category = categorize_function(filename)
            target_dir = os.path.join("/workspaces/finetune/data_test_labels", category)
            if not os.path.exists(target_dir):
                os.makedirs(target_dir)
            shutil.move(os.path.join(target_dir, filename), os.path.join(target_dir, filename))
            print(f"Moved {filename} to {target_dir}/")

# Set the directory where your images are currently stored
source_directory = "/workspaces/finetune/data_test"
organize_images(source_directory)
```

```python
import os
import shutil

def move_folders_to_new_parent(source_dir, parent_folder_name):
    # Path to the new parent folder
    parent_folder_path = os.path.join(source_dir, parent_folder_name)

    # Create the parent folder if it does not exist
    if not os.path.exists(parent_folder_path):
        os.makedirs(parent_folder_path)
        print(f"Created new directory: {parent_folder_path}")

    # Iterate over each item in the source directory
    for item in os.listdir(source_dir):
        item_path = os.path.join(source_dir, item)
        
        # Check if the item is a directory and not the new parent folder
        if os.path.isdir(item_path) and item != parent_folder_name:
            # Move the directory to the new parent folder
            shutil.move(item_path, parent_folder_path)
            print(f"Moved {item_path} to {parent_folder_path}/")

# Set the directory where your images are currently stored
source_directory = "/workspaces/finetune/data_test"
new_parent_folder = "organized_categories"

move_folders_to_new_parent(source_directory, new_parent_folder)
```
